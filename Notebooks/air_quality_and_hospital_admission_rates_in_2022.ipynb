# Focusing on Data from Kentucky

```python
#Boolean indexing to drop other states and focus on Kentucky
ky_aqi_boolean = annual_aqi['State'] == 'Kentucky'
ky_aqi_boolean
```

```python
#Drop false values from the prior boolean indexing
ky_aqi_cleaned = ky_aqi_boolean.drop(annual_aqi[ky_aqi_boolean == False].index)
ky_aqi_cleaned
#Shows that KY has 26 counties with AQI data in 2022, rows 317 to 342
```

```python
#Callback to annual_aqi to get KY data
ky_aqi = annual_aqi[annual_aqi['State'] == 'Kentucky']
ky_aqi
```

Some of these columns don't really have substantial data, such as the Very Unhealthy Days and the Hazardous Days columns. We'll drop the mostly empty columns since we're focusing on gases/PM values.

```python
#Drop columns that aren't necessary or lack sufficient data
ky_aqi = ky_aqi.drop(columns=['Unhealthy for Sensitive Groups Days', 'Unhealthy Days', 'Very Unhealthy Days', 'Hazardous Days', 'Days CO', 'Days NO2'])
ky_aqi
```

Let's narrow all of this down to regional data:
- Fayette (North) <br>
- Pulaski (South) <br>
- Pike (East) <br>
- Warren (West)

```python
#Narrowing down the counties to Fayette, Pulaski, Pike, and Warren
selected_counties = ['Fayette', 'Pulaski', 'Pike', 'Warren']
ky_selected_counties = ky_aqi[ky_aqi['County'].isin(selected_counties)]
ky_selected_counties
```

Now, specifically, I want to look at days with PM2.5(Air pollution) and Ozone. So I'll drop other columns to make the graphs.

```python
#Taking off columns besides PM2.5 and Ozone
ky_ozone = ky_selected_counties.drop(columns=['Days PM10', 'Days PM2.5', 'Median AQI', '90th Percentile AQI', 'Max AQI', 'Moderate Days', 'Good Days', 'Days with AQI', 'Year', 'State'])

ky_pm25 = ky_selected_counties.drop(columns=['Days PM10', 'Days Ozone', 'Median AQI', '90th Percentile AQI', 'Max AQI', 'Moderate Days', 'Good Days', 'Days with AQI', 'Year', 'State'])
```

```python
ky_ozone.head()
```

```python
ky_pm25.head()
```

# Making the Visualizations

```python
plt.figure(figsize=(10,8))
plt.bar(ky_ozone['County'], ky_ozone['Days Ozone'], edgecolor='black', linewidth=1, alpha=1, color='#0072B2')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.ylim(0, 365)

plt.title('Number of Ozone Days in 2022 for Selected Kentucky Counties (Annually)', fontsize=16, pad=20)
plt.xlabel('County', fontsize=14)
plt.ylabel('Number of Ozone Days (Annually)', fontsize=14)
plt.xticks(rotation=45)

plt.savefig('../Visualizations/ozone_days_ky_counties_2022.jpg')
plt.show()
```

The above chart shows that, out of 365 days in the year, the amount of Ozone Days is actually under half for these four selected counties. 
- I chose blue because it seemed right for Ozone and is generally colorblind friendly <br>
- I chose a bar chart because I'm comparing the amount of days of Ozone between four counties <br>

Now we'll look at PM2.5.

```python
plt.figure(figsize=(10,8))
plt.bar(ky_pm25['County'], ky_pm25['Days PM2.5'], edgecolor='black', linewidth=1, alpha=1, color='#D55E00')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.ylim(0, 365)

plt.title('Number of PM2.5 Days in 2022 for Selected Kentucky Counties (Annually)', fontsize=16, pad=20)
plt.xlabel('County', fontsize=14)
plt.ylabel('Number of PM2.5 Days (Annually)', fontsize=14)
plt.xticks(rotation=45)

plt.savefig('../Visualizations/pm25_days_ky_counties_2022.jpg')
plt.show()
```

This data shows us that the PM2.5 values in some counties did hit over half of the 365 days in a year.

- I chose another bar chart because I'm still comparing by county, but now with PM2.5 values<br>
- I chose orange because it feels right for pollen and is generally considered a good contrast to blue for colorblind folks<br>

Now let's see these values side by side.

```python
plt.figure(figsize=(10,8))

w, x= 0.3, np.arange(len(ky_ozone['County']))

fig, ax = plt.subplots()
ax.bar(x - w/2, ky_ozone['Days Ozone'], width=w, label='Ozone Days', color='#0072B2', edgecolor='black', linewidth=1)
ax.bar(x + w/2, ky_pm25['Days PM2.5'], width=w, label='PM2.5 Days', color='#D55E00', edgecolor='black', linewidth=1)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.ylim(0, 365)

plt.title('Ozone vs. PM2.5 Days in 2022 for Selected Kentucky Counties (Annually)', fontsize=16, pad=20)
plt.xlabel('County', fontsize=14)
plt.ylabel('Number of Days (Annually)', fontsize=14)
plt.xticks(x, ky_ozone['County'], rotation=45)
plt.legend()


plt.savefig('../Visualizations/ozone_vs_pm25_days_ky_counties_2022.jpg', bbox_inches='tight')
plt.show()
```

Altogether it looks like the PM2.5 days far outnumber the Ozone days. Neither option is great for anyone with pulmonary issues, but those pesky PM2.5 days definitely made it harder for those of us in the selected counties in 2022.

- The compounded bar chart gives an idea of air quality annually when put together side-by-side<br>

Focusing on data from Kentucky

```python
#Boolean indexing to drop other states and focus on Kentucky
ky_aqi_boolean = annual_aqi['State'] == 'Kentucky'
ky_aqi_boolean
```

```python
#Drop false values from the prior boolean indexing
ky_aqi_cleaned = ky_aqi_boolean.drop(annual_aqi[ky_aqi_boolean == False].index)
ky_aqi_cleaned
#Shows that KY has 26 counties with AQI data in 2022, rows 317 to 342
```

```python
#Callback to annual_aqi to get KY data
ky_aqi = annual_aqi[annual_aqi['State'] == 'Kentucky']
ky_aqi
```

Some of these columns don't really have substantial data, such as the Very Unhealthy Days and the Hazardous Days columns. We'll drop the mostly empty columns since we're focusing on gases/PM values.

```python
#Drop columns that aren't necessary or lack sufficient data
ky_aqi = ky_aqi.drop(columns=['Unhealthy for Sensitive Groups Days', 'Unhealthy Days', 'Very Unhealthy Days', 'Hazardous Days', 'Days CO', 'Days NO2'])
ky_aqi
```
